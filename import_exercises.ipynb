2. Read about and use the itertools module from the python standard library to help you solve the following problems:

How many different ways can you combine the letters from "abc" with the numbers 1, 2, and 3?

```python
from itertools import permutations 
  
  
a = "abc123"
  
# no length entered so default length
# taken as 6(the length of string 'abc123')
p = permutations(a)

# adding a count feature

count = 0
  
# Print the obtained permutations 
for j in list(p):
    count += 1
    print(count)
    
# I got 720 possible permutations from the string'abc123'
# which checks out based on the mathematical check of 6! = 720
    
```

```python
ft = list(itertools.permutations('abc123'))
ft
count = 0
for i in ft:
    count += 1
print(count)
```

```python
# Most common favorite fruit

ff = [profile['favoriteFruit'] for profile in profiles]

# make a dict to count values of fruits

fruit_counts = {}

for fruit in ff:
    
# creates a key in dict if new fruit and starts count at 1    
    if fruit not in fruit_counts.keys():
        fruit_counts[fruit] = 1
# key already exists in dict, so we add 1 to the count
    else:
        fruit_counts[fruit] += 1
# dict of fruits with keys and counts
fruit_counts
```

```python
# getting a printout of the most popular
for k, v in fruit_counts.items():
    if v == max(fruit_counts.values()):
        print(k)
```

```python
# getting a printout of the least most popular
for k, v in fruit_counts.items():
    if v == min(fruit_counts.values()):
        print(k)
```
